```python
def drop_above_below_percentil(df, column, above_below, percentil):
    if (above_below != 'above') & (above_below != 'below'):
        raise TypeError('Value MUST be "above" or "below"')
    
    q = df[column].quantile(percentil)
    print(q)
    if above_below == 'above':
        df = df[df[column]<q]
        return df
    elif above_below == 'below':
        df = df[df[column]>q]
        return df
        
percentil_df = drop_above_below_percentil(short_df, 'price', 'below', 0.95)
percentil_df['price'].count()
    
```

```python
def outlier(df, column):
    q1 = df[column].quantile(0.25)
    print(q1)
    q3 = df[column].quantile(0.75)
    print(q3)
    iqr = q3 - q1

    lower_bound = q1 -(1.5 * iqr) 
    upper_bound = q3 +(1.5 * iqr)

    df = df.drop(df[df[column] < lower_bound].index)
    df = df.drop(df[df[column] > upper_bound].index)
            
    
    return(df)
```
